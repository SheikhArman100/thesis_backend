```python
import pandas as pd
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import string
```

```python
def clean_stopword_punc(dataset,column):
    rows = len(dataset)
    count = 0
    stop_words = set(stopwords.words('english'))
    punc = list(string.punctuation)
    while count < rows:
        column_details = dataset.at[count, column]
        words = word_tokenize(column_details)
        filtered_sentence = []
        for w in words:
            if w not in punc:
                if w not in stop_words:
                    filtered_sentence.append(w)
        new_column_details = ''

        new_column_details = ' '.join(map(str, filtered_sentence))
        dataset.loc[count, column] = new_column_details

        count += 1
    return dataset
clean_stopword_punc(dataset,'description')
dataset.head()
```

```python
from nltk.stem import PorterStemmer
def stemming(dataset,column):
  stemmer = PorterStemmer()
  rows = len(dataset)
  count = 0
  while count < rows:
        column_details = dataset.at[count, column]
        words = word_tokenize(column_details)
        filtered_sentence = []
        for word in words:
          updated_word=stemmer.stem(word)
          filtered_sentence.append(updated_word)
        new_column_details = ''
        new_column_details = ' '.join(map(str, filtered_sentence))
        dataset.loc[count, column] = new_column_details
        count+=1
  return dataset

stemming(dataset,'description')
dataset.head()






```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

def lemmatization(dataset,column):
  lemmatizer = WordNetLemmatizer()
  rows = len(dataset)
  count = 0
  while count < rows:
    column_details = dataset.at[count, column]
    words=word_tokenize(column_details)
    filtered_sentence = []
    for word in words:
          updated_word=lemmatizer.lemmatize(word)
          filtered_sentence.append(updated_word)
    new_column_details = ''
    new_column_details = ' '.join(map(str, filtered_sentence))
    dataset.loc[count, column] = new_column_details
    count+=1
  return dataset

lemmatization(dataset,'description')
dataset.head()




```

```python

keywords = []
def text_rank(dataset,column):
  nlp = spacy.load('en_core_web_sm')
  nlp.add_pipe("textrank")
  rows = len(dataset)
  count = 0
  while count < rows:
    column_details = dataset.at[count, column]
    doc = nlp(column_details)
    graph, rank = pytextrank.text_rank(doc)
    kw = pytextrank.get_key_phrases(graph)
    keywords.append(kw)
    count+=1
  return keywords
text_rank(dataset,'description')




```

```python
def withoutPunctuation(dataset,column):

   line = " "
   for word in dataset[column]:
     line = line +  ' ' + word

   a = line

   fwords = nltk.word_tokenize(line)

   fwords = [word.lower() for word in fwords if word.isalpha()]

   for word in fwords:
      notPunc[word.lower()] = notPunc[word.lower()] + 1

   notPunc10= notPunc.most_common()

   return notPunc10

withoutPunctuation(dataset,'description')
```

```python

import nltk
from nltk.tokenize import word_tokenize

def search_rows(dataset,word_column,modified_column,keywords):
  rows = len(dataset)
  count = 0
  while count < rows:
        description = dataset.at[count, word_column]
        words = word_tokenize(description)
        for word in words:
          for keyword in keywords:
            if word==keyword:
              print(f"Keyword found in row {count}")
              dataset.loc[count,modified_column]=keyword







        count+=1
  return dataset


```

```python
#getting  car_owner and owner_info from shop_info

from nltk.tokenize import sent_tokenize
dataset['shop_info'] = dataset['shop_info'].str.replace("\n", ". ")

def shop_info_details(dataset):
   rows=len(dataset)
   count=0
   while count < rows:
      shopInfo = dataset.at[count, 'shop_info']
      sentences = sent_tokenize(shopInfo)
      for sentence in sentences:
        sentence=sentence.replace('.','',)
      if sentences[0]=='For sale by.':
        dataset.at[count,'car_owner']=sentences[1]
        try:
          if sentence[2]=="MEMBER.":
            dataset.loc[count,'owner_info']=sentences[2].replace('.','')
          else:
            dataset.at[count,'owner_info']=None
        except :
          dataset.at[count,'owner_info']=None



      else:
        dataset.at[count,'car_owner']=sentences[0].replace('.','')
        dataset.loc[count,'owner_info']=sentences[1].replace('.','')

      count+=1
   return dataset

shop_info_details(dataset)

```
